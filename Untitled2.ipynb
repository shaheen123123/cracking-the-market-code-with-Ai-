```python
# ✅ STEP 1: Install required libraries
!pip install yfinance --quiet

# ✅ STEP 2: Import libraries
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# ✅ STEP 3: Load stock data
ticker = "AAPL"  # You can change this to any valid stock symbol
df = yf.download(ticker, start="2015-01-01", end="2024-12-31")
df = df[["Close"]]
df.dropna(inplace=True)

# ✅ STEP 4: Visualize closing prices
plt.figure(figsize=(12,6))
plt.plot(df["Close"], label="Closing Price")
plt.title(f"{ticker} Closing Price History")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.legend()
plt.show()

# ✅ STEP 5: Prepare data
data = df.values
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

# Create sequences
def create_dataset(dataset, time_step=60):
    X, y = [], []
    for i in range(time_step, len(dataset)):
        X.append(dataset[i-time_step:i])
        y.append(dataset[i])
    return np.array(X), np.array(y)

time_step = 60
X, y = create_dataset(scaled_data, time_step)
X = X.reshape((X.shape[0], X.shape[1], 1))

# Split into train/test sets
split = int(len(X) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# ✅ STEP 6: Build LSTM model
model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    LSTM(50),
    Dense(1)
])
model.compile(optimizer="adam", loss="mean_squared_error")
model.summary()

# ✅ STEP 7: Train model
model.fit(X_train, y_train, epochs=10, batch_size=32)

# ✅ STEP 8: Make predictions
predicted = model.predict(X_test)
predicted_prices = scaler.inverse_transform(predicted)
actual_prices = scaler.inverse_transform(y_test)

# ✅ STEP 9: Plot predictions
plt.figure(figsize=(12,6))
plt.plot(actual_prices, label="Actual Price")
plt.plot(predicted_prices, label="Predicted Price")
plt.title(f"{ticker} Price Prediction (LSTM)")
plt.xlabel("Time")
plt.ylabel("Price ($)")
plt.legend()
plt.show()

```
